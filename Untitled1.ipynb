```python
cols = ['operation', 'property_type', 'place_name', 'place_with_parent_names',
       'country_name', 'geonames_id', 'lat-lon', 'lat', 'lon', 'price',
       'currency', 'price_aprox_local_currency', 'price_aprox_usd',
       'surface_total_in_m2', 'surface_covered_in_m2', 'price_usd_per_m2',
       'price_per_m2','properati_url','description', 'title', 'image_thumbnail', 'floor', 'rooms','expenses']
df2 = df.drop(cols, axis=1)
```

```python
# Seleccionar modelo
from sklearn.linear_model import LinearRegression
# Elegir hiperparmetros
model = LinearRegression(fit_intercept=True)
# Preparar los datos en una matriz de features
# Crear X e y
X = df2 #2 dimensiones
y = df['price'] #1 dimensin

# Split entrenamiento / testeo para CV
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=1)
#Ajustar el modelo a los datos
model.fit(Xtrain, ytrain)
print (model.coef_)
print (model.intercept_)
```

```python
#Evaluar
ypred = model.predict(Xtest)
from sklearn import metrics
print ('MAE:', metrics.mean_absolute_error(ytest, ypred))
print ('MSE:', metrics.mean_squared_error(ytest, ypred))
print ('RMSE:', np.sqrt(metrics.mean_squared_error(ytest, ypred)))
print ('R2:', metrics.r2_score(ytest, ypred))
```
